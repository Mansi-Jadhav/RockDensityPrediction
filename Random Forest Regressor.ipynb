```python
# Linear Regression is not suitable for sine wave data
# It calculates based on errors from points
# It gives the same results for varying data
```

```python
def run_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_preds)
    rmse = np.sqrt(y_test, y_preds)
    print(f'MAE: {mae}')
    print(f'RMSE: {rmse}')
    
    signal_range = np.arange(0,100).reshape(-1,1)
    signal_preds = model.predict(signal_range)

    plt.figure(figsize=(5,4), dpi=100)
    sns.scatterplot(x='Signal', y='Density', data=df)
    plt.plot(signal_range, signal_preds, color='red')
    plt.show()
```

```python
lr_model = LinearRegression()
run_model(lr_model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
```

```python
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=4), LinearRegression())
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=6), LinearRegression())
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=10), LinearRegression())
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
# As we keep increasing the degree, model is overfit to the training data
# If we know for sure, our values are always gonna be between 0 to 100, maybe we could have used this model
# But ideally, it's not the best choice as it doesn't correctly predict the output
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k_values=[1,5,10]  # We're gonna test 3 different models

for k in k_values:
    model = KNeighborsRegressor(n_neighbors=k)
    run_model(model, X_train, y_train, X_test, y_test)
```

```python
# Noisy and more biased
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr = SVR()
param_grid = {'C':[0.01,0.1,1,5,10,100,1000],'gamma':['auto','scale']}
model = GridSearchCV(svr, param_grid)
run_model(model, X_train, y_train, X_test, y_test)
```

```python
# Fits quiet well
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=10)
run_model(rfr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
model = GradientBoostingRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```
